# Analysis
This file provides tables for analysis from the data collected in wikipedia.ipynb. There are 6 

```python
import pandas as pd

df = pd.read_csv('wp_scored_city_articles_by_state.csv')
df.head()
```

1. Top 10 US states by coverage: The 10 US states with the highest total articles per capita (in descending order) .


```python
df_grouped = df.groupby('state')
# Specify aggregation functions for 'Value1' and 'Value2' columns
aggregations = {'article_title': 'count', 'population': 'max'}

# Apply the aggregation functions to each group
grouped_result = df_grouped.agg(aggregations)

grouped_result['per capita articles'] = grouped_result['article_title']/grouped_result['population']
grouped_result = grouped_result.rename(columns={'article_title': 'articles'})

grouped_result.sort_values(by='per capita articles', ascending=False).head(10)
```

2. Bottom 10 US states by coverage: The 10 US states with the lowest total articles per capita (in ascending order) .

```python
grouped_result.sort_values(by='per capita articles', ascending=True).head(10)
```

3. Top 10 US states by high quality: The 10 US states with the highest high quality articles per capita (in descending order) .

```python
filtered_rows = df[(df['article_quality'] == 'FA') | (df['article_quality'] == 'GA')]
df_grouped = filtered_rows.groupby('state')
# Specify aggregation functions for 'Value1' and 'Value2' columns
aggregations = {'article_title': 'count', 'population': 'max'}

# Apply the aggregation functions to each group
grouped_result = df_grouped.agg(aggregations)
grouped_result['per capita good articles'] = grouped_result['article_title']/grouped_result['population']
grouped_result = grouped_result.rename(columns={'article_title': 'good articles'})


grouped_result.sort_values(by='per capita good articles', ascending=False).head(10)
```

4. Bottom 10 US states by high quality: The 10 US states with the lowest high quality articles per capita (in ascending order).


```python
grouped_result.sort_values(by='per capita good articles', ascending=True).head(10)
```

5. Census divisions by total coverage: A rank ordered list of US census divisions (in descending order) by total articles per capita.

```python
df_grouped = df.groupby('state')
# Specify aggregation functions for 'Value1' and 'Value2' columns
aggregations = {'article_title': 'count', 'population': 'max', 'regional_division':'first'}

# Apply the aggregation functions to each group
grouped_result = df_grouped.agg(aggregations)

aggregations = {'article_title': 'sum', 'population': 'sum'}
df_grouped2 = grouped_result.groupby('regional_division')
grouped_result2 = df_grouped2.agg(aggregations)


grouped_result2['per capita good articles'] = grouped_result2['article_title']/grouped_result2['population']
grouped_result2 = grouped_result2.rename(columns={'article_title': 'good articles'})


grouped_result2.sort_values(by='per capita good articles', ascending=False).head(10)
```
